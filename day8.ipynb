```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
# EDA , Univariant
# categorical data
# count plot

plt.figure(figsize=(6,4))
sns.countplot(x='Survived', data=df)
plt.show()
df['Survived'].value_counts()
```

```python

plt.figure(figsize=(6,4))
sns.countplot(x='Pclass', data=df)
plt.title("Count of Passenger Class")
plt.show()

df['Pclass'].value_counts()
```

```python
df['Pclass'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
df['Sex'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
# Numerical Data
plt.hist(df['Age'])
```

```python
sns.displot(df['Age'])
```

```python
sns.boxplot(df['Age'])
```

```python
sns.boxplot(df['Fare'])
```

```python
# univariant , multivariant
# 1. numerical-numeriacal.
# Scatter plot
sns.scatterplot(x='Age', y='Fare', data=df, hue='Sex',style='Survived')
```

```python
# Bar plot ( numerical- categorical )

sns.barplot(x= 'Pclass', y='Fare', data=df)
```

```python
# categorical-categorical
#headplot
df.head(3)
```

```python
sns.heatmap(pd.crosstab(df[ 'Pclass'],df['Survived']))
```

```python
sns.clustermap(pd.crosstab(df[ 'SibSp'],df['Survived']))
```

```python
sns.pairplot(df, hue='Pclass')
```

```python
#line plot ( numerical - numerical) (time basesd - date, age ,year)
new = df.groupby('Age').sum().reset_index()
```

```python
sns.lineplot(x='Age', y='Fare',data=new)
```
